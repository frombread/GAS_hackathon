```python
x_data = df[["65세 이상 인구수","장애인구","기초생활수급자인구수"]]
target = df[["연체미납03월"]]
x_data1 = sm.add_constant(x_data, has_constant = "add")
multi_model = sm.OLS(target, x_data1)
fitted_multi_model = multi_model.fit()
fitted_multi_model.summary()
```

```python
x_data = df[["65세 이상 인구수","장애인구"]]
target = df[["연체미납03월"]]
x_data1 = sm.add_constant(x_data, has_constant = "add")
multi_model = sm.OLS(target, x_data1)
fitted_multi_model = multi_model.fit()
fitted_multi_model.summary()
```

```python
x_data = df[["65세 이상 인구수"]]
target = df[["연체미납03월"]]
x_data1 = sm.add_constant(x_data, has_constant = "add")
multi_model = sm.OLS(target, x_data1)
fitted_multi_model = multi_model.fit()
fitted_multi_model.summary()
```

```python
from scipy.stats import probplot

fig = plt.figure(figsize=(8,8))
fig.set_facecolor('white')
ax = fig.add_subplot()
probplot(fitted_multi_model.resid, dist='norm',plot=ax) 
plt.show()
```
